```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 3
n = 5

M = np.random.randint(10,size=(m, n))
print(M)

plt.plot(M)
```

```python
print(M)
plt.imshow(M)

for i in range(m):
    for j in range(n):
        plt.text(j, i, str(M[i, j]), 
                 fontsize=20, 
                 horizontalalignment='center', 
                 verticalalignment='center') # the x and y coordinates are flipped in text function

plt.colorbar()
plt.show()
```

```python
# Exercise: visualise hilbert matrix using seaborn (heatmap)
# H is a 10x10 matrix

import numpy as np
import seaborn as sns

H = np.zeros((10, 10))

for i in range(10):
    for j in range(10):
        H[i,j]=1/((i+1)+(j+1)-1)

sns.heatmap(H,vmax = 0.4)
plt.show()
```

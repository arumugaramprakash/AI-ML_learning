```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# low resolution image when zoomed in looks pixelated.
x = np.linspace(.5,5,10)
y1 = np.log(x)
y2 = 2-np.sqrt(x)

plt.plot(x,y1,'bo-', label = 'log')
plt.plot(x,y2,'rs-', label = 'sqrt')
plt.legend()

plt.show()
```

```python
# high resolution image when zoomed in looks nice.
from IPython import display
display.set_matplotlib_formats('svg')

x = np.linspace(.5,5,10)
y1 = np.log(x)
y2 = 2-np.sqrt(x)

plt.plot(x,y1,'bo-', label = 'log')
plt.plot(x,y2,'rs-', label = 'sqrt')
plt.legend()


plt.savefig('test.png') # saves the figure as a png file can also be saved as pdf or svg just by changing the extension.
plt.show()
```

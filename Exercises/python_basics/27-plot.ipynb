```python
import matplotlib.pyplot as plt
plt.plot(3,4,'ro')
# o is for dot
# x is for x
# s is for square
# ^ is for triangle up
# v is for triangle down
# r is for red color
#  g is for green color
#  b is for blue color
#  y is for yellow color
#  k is for black color
#  w is for white color
```

```python
plt.plot(3,4,'r^')
plt.plot(2,4.2,'bo')
```

```python
plt.plot(3,4,'r^')
plt.show() # ends all of the previous plt.plots done so far and shows them in a single window
plt.plot(2,4.2,'bo')
```

```python
plt.plot(3,4,'r^',label='red triange')
plt.plot(2,4.2,'bo',label='blue dot')
plt.legend() # legend is created on the order the plt.plot is created.
plt.show()
```

```python
plt.plot(3,4,'r^')
plt.plot(2,4.2,'bo')
plt.legend(['red','blue']) # Be midful on the order as it connects with the oreder of the py plot lib
plt.show()
```

```python
x = [0,1]
y = [0,2]

# this is similare to p1 =(0,0) and P2 = (1,2) and plot a line from p1 to p2
plt.plot(x,y,'m') # m stands for magenta color
plt.show()

```

```python
x = [0,1,3]
y = [0,2,-1]

plt.plot(x,y,'m+:')
plt.show()

# m is for magenta line
# ms give a magenta square marker without line
# ms- gives a magenta square marker with line
# b-- gives blue dashed line
# r+- gives red plus markers with lines
# g: gives green dotted line
# k.- gives black dot and dash line
# o- gives orange circle markers with lines
# x- gives cross markers with lines

```

```python
import numpy as np
x = np.linspace(0,3*np.pi,101)
y = np.sin(x)

plt.plot(x,y,'m-')
# plt.axis('square')
plt.show()
```

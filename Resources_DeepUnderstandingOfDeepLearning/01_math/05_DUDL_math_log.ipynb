```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ploting a exponent
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2,2,1000)
a = np.exp(x)
b = np.log(np.exp(x))
c = np.log(x)

plt.plot(x, a, 'r-', label = '$e^x$')
plt.plot(x,b,'g-', label = '$e^{log(x)}$')
plt.plot(x,c,'b-', label = '$log(x)$')

plt.legend()
plt.xlabel('x')
plt.ylabel('fn')
plt.show()

```

```python
# define a set of points to evaluate
x = np.linspace(.0001,1,200)

# compute their log
logx = np.log(x)



# plot!
fig = plt.figure(figsize=(10,4))

# increase font size. FYI
plt.rcParams.update({'font.size':15})

plt.plot(x,logx,'ks-',markerfacecolor='w')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

```python
# demonstration that log and exp are inverses

# redefine with fewer points
x = np.linspace(.0001,1,20)

# log and exp
logx = np.log(x)
expx = np.exp(x)

# the plot
plt.plot(x,x,color=[.8,.8,.8])
plt.plot(x,np.exp(logx),'o',markersize=8)
plt.plot(x,np.log(expx),'x',markersize=8)
plt.xlabel('x')
plt.ylabel('f(g(x))')
plt.legend(['unity','exp(log(x))','log(exp(x))'])
plt.show()
```

```python
# Plotting natural log, log of base 2 and log base 10
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2, 2, 1000)
a = np.log(x)
b = np.log2(x)
c = np.log10(x)

plt.plot(x, a, 'r-', label='$\ln(x)$')
plt.plot(x, b, 'g-', label='$log_2(x)$')
plt.plot(x, c, 'b-', label='$log_{10}(x)$')

plt.legend()
plt.xlabel('x')
plt.ylabel('fn')
plt.show()
```
